```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("data.xlsx")
df.head()
```

```python
plt.hist(df['Цена'], bins=20, color='blue', edgecolor='black')
plt.xlabel("значения")
plt.ylabel("частота")
plt.title("гистограмма распределения")
plt.show()
```

```python
plt.hist(df['Выручка  '], bins=10, color='blue', edgecolor='black')
plt.xlabel("выручка")
plt.ylabel("частота")
plt.title("распределение выручки")
plt.show()
```

```python
df["Дата"]=pd.to_datetime(df["Дата"])
df.sort_values("Дата", inplace=True)
```

```python
plt.plot(df["Дата"], df["Выручка  "], marker='o', linestyle='-',color='green')
plt.xlabel("Дата")
plt.ylabel("Выручка")
plt.title("Динамика выручки по дням")
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
df_grouped=df.groupby("Товар")["Количество"].sum()
df_grouped.plot(kind='bar', color = 'orange', edgecolor='black')
plt.xlabel("Товар")
plt.ylabel("Кол-во проданных единиц")
plt.title("Общее кол-во продаж по товарам")
plt.show()
```

```python
print("Средняя выручка: ", df["Выручка  "].mean())
print("Медианная выручка: ", df["Выручка  "].median())
print("Стандартное отклонение", df["Выручка  "].std())
```
